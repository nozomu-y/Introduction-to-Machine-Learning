Multiple classifiers can solve the same problem.  


## Load data

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

X is the features and y is the labels.  
We can think of a classifier as the following function.  
\begin{equation}
    f(X) = y
\end{equation}

## Partition data into two parts
In order to test how accurate our model is, we partition our data into two parts, train and test.  


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .5)   # Use half the data for testing
```

## Using different classifiers
By replacing the following 2 lines, we can use a different classifier.  

### Decision tree

```python
from sklearn import tree
my_classifier = tree.DecisionTreeClassifier()
```

### K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
my_classifier = KNeighborsClassifier()
```

```python
my_classifier.fit(X_train, y_train)
predictions = my_classifier.predict(X_test)
print(predictions)
```

## Calculate the accuracy of the classifier

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions))
```
