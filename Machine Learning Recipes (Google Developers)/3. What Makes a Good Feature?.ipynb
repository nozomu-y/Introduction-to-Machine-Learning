Classifiers can be only as good as the features that are provided.   
Thus, coming up with good features is one of the most important jobs in ML.  

## Classifying two types of dogs (greyhounds and labradors)

Let's assume the following. 
* Greyhounds are usually taller than labradors
* Dogs have only two eye colors (blue and brown)
* The color of the eyes doesn't depend on the breed of dog

Some features are useful, while others tell us nothing.  
In this case, height are a useful feature, but the eye color isn't useful at all. 


Greyhounds tends to be a couple of inches taller than labradors, but not always.   
When we think of features, we have to consider how it looks for different values in a population. 

```python
import numpy as np 
import matplotlib.pyplot as plt 

greyhounds = 500
labs = 500

grey_height = 28 + 4 * np.random.randn(greyhounds)  # inches
lab_height = 24 + 4 * np.random.randn(labs)

plt.hist([grey_height, lab_height], stacked=True, color=['r', 'b'], label=['Greyhound', 'Labrador'])
plt.xlabel("Height")
plt.ylabel("Number of dogs")
plt.legend()
plt.show()
```

The above shows how many dogs in our population have a given height.  
In the middle of the histogram, the probability of each type of dog is close.  
Thus, height is a useful feature, but it's not perfect.  
This is why we need multiple features in ML.  


### Exclude useless features
Since eye color doesn't depend on the type of dog, it tells us nothing.  
Including a useless feature like this in the training dataset can decrease the accuracy.  


### Make your features independent
Independent features gives you different types of information.  
Features such as 
* Height in inches 
* Height in centimeters  

are perfectly correlated with one another.  
So we should remove such data.  
This is because a lot of classifiers are not smart enough to realize that these are the same thing, and they might doble count how important this features is.  

### Features should be easy to understand
Distance between the cities and coordinates for each cities can both express how far apart they are.  
Though, the distance between the cities is way more easier to understand.  
Using coordinates is much harder and would require many more examples in your training data. 


## Ideal features are...
* Informative
* Independent
* Simple

There are techniques to figure out exactly how useful your features are, and even what combinations of them are the best.
