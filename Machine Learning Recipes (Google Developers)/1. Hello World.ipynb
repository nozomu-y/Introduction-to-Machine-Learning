## Classifying apples and oranges

Writing codes that can tell the difference between an apple and an orange is impossible without using machine learning.  
**Supervised Learning**: Create a classifier by finding patterns in examples. 

### Training data
| Weight | Texture | Label |
| ---- | ---- | ---- |
| 150g | Bumpy | Orange | 
| 170g | Bumpy | Orange | 
| 140g | Smooth | Apple | 
| 130g | Smooth | Apple |

**Feature**: an individual measurable property or characteristic of a phenomenon being observed. (Weight and Texture)  
**Target**: Final output you are trying to predict.  
**Label**: True outcome of the target. This is known for the training dataset but not for the test.  

```python
from sklearn import tree
```

```python
features = [[140, 1], [130, 1], [150, 0], [170, 0]]
```

Since scikit-learn uses real-valued features, we will use 0 for bumpy and 1 for smooth.  

```python
labels = [0, 0, 1, 1]
```

Likewise, 0 for apples and 1 for oranges. 

```python
clf = tree.DecisionTreeClassifier()
clf.fit(features, labels)
```

```python
print(clf.predict([[150, 0]]))
```

The classifier predicted that a 150g, bumpy fruit is an orange.
